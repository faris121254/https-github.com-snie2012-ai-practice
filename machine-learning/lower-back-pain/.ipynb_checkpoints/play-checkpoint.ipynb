```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
import re
```

```python
df = pd.read_csv('../data/lower-back-pain-symptoms/Dataset_spine.csv')
```

```python
keys = df['Unnamed: 13'].dropna().apply(lambda x : x.split('='))
keys = keys[keys.map(len) > 1].map(lambda x: [x[0].strip(), x[1].strip()])
keys = keys.map(lambda x: [int(re.search('(\d+)', x[0]).group()), x[1]])
```

```python
cols = keys.sort_values().apply(lambda x: x[1]).append(pd.Series('class'))
```

```python
cols = cols.apply(lambda x: x.split('(')).apply(lambda x: x[0].strip())
```

```python
df = df.drop(['Unnamed: 13'], axis=1)
df.columns = cols
```

```python
df.sample(10)
```

```python
groups = df.groupby('class')
```

```python
gs = {}
for key, g in groups:
    gs[key] = g
```

```python
normal = gs['Normal'].describe()
abnormal = gs['Abnormal'].describe()
```
